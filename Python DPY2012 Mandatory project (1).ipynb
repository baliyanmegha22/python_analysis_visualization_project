```python
#importing library 
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
#reading excel file using pandas excel command
df = pd.read_excel('housing.xlsx')
```

```python
#converting it into dataframe
data_file = pd.DataFrame(df)
data_file
```

Bar chart are useful but here we have a lot of data then we will have 100 different column so we use histogram as it uses bins for values. 

```python
# 1. What is the average median income of the data set and check the distribution of data using appropriate plots. 
# Please explain the distribution of the plot

#average median income calculation using dataframe. 
avg_median_income = df['median_income'].mean()
print("Average Median Income:", avg_median_income)
plt.style.use('fivethirtyeight') # change the style or appearance of a plot
#plotting histogram to show frequency distribution of the median income values
plt.hist(df['median_income'], bins=25, density=True, color='green', edgecolor='black')
sns.kdeplot(df['median_income'], color='green', linewidth=2, shade=True)
plt.xlabel('Median Income')
plt.ylabel('Density')
plt.title('Distribution of Median Income',color ='orange')
plt.show()

# displays the frequency of different income ranges, and a smooth curve called a kernel density plot that estimates 
#the density of income values.Insights: significant gaps between different income levels.
```

```python
# 2. Draw an appropriate plot to see the distribution of housing_median_age and explain your observations.
plt.style.use('fivethirtyeight') # change the style or appearance of a plot
plt.hist(df['housing_median_age'], bins=15, density=True, width = 2,color = 'blue',edgecolor ='black')
sns.kdeplot(df['housing_median_age'], color='blue', linewidth=2, shade=True)
plt.xlabel('housing_median_age')
plt.ylabel('Density')
plt.title('Distribution of Housing Median age',color = 'red')
plt.show()

#from the graph we can most common housing median age range by identifying the peak or highest point. 
#A wider spread indicates a broader range of ages in the dataset, while a narrower spread suggests a more focused 
# distribution. It suggests that the majority of houses in the dataset tend to have median ages that are either 
# predominantly younger or predominantly older.
```

```python
# 3. Show with the help of visualization, how median_income and median_house_values are related?

median_income_data = df['median_income']
median_house_values_data = df['median_house_value']

# Observation: Areas with higher point density suggest a larger number of data points with similar income
plt.style.use('fivethirtyeight') # change the style or appearance of a plot
plt.scatter(median_income_data, median_house_values_data, marker = 's', edgecolor = 'c', facecolor ='k' , s = 0.7)
plt.xlabel('Median Income')
plt.ylabel('Median House Value')
plt.title('Relationship between Median Income and Median House Values')
plt.show()

# higher income levels are associated with higher house values and the average median house value is low as below <300000
# as well as the average of median income. Hgher income levels also ssociated with lower house values.
```

```python
# 4. Create a data set by deleting the corresponding examples from the data set for which total_bedrooms are not available

#temp = df.isnull().sum() #to check null values in column
#temp

#dealing with missing value by deleting the empty data from the column total_bedrooms
dataset_without_missing = df.dropna(subset=['total_bedrooms']) #as we had null value in a column using subset to use 
#column name to remove the nan value. 
dataset_without_missing
```

```python
# 5. Create a data set by filling the missing data with the mean value of the total_bedrooms in the original data set.

#handling null value by filling the data using fillna function and provided the value of mean value of the total_bedrooms 
#for the column total bedrooms(taking as key) and the mean as value using dctionary. 
 
missing_data = df.fillna({'total_bedrooms': df['total_bedrooms'].mean()})
missing_data
```

```python
#6. Write a programming construct (create a user defined function) to calculate the median value of the data set 
#wherever required
def cal_median(df): # DataFrame as input and calculates the median value using the NumPy library
    return np.median(df) #median of the DataFrame using the NumPy library's median() function.The calculated median value 
#is then returned as the output of the function.

# Example usage
median_value = cal_median(df['median_income']) #calling function
print("Median Value:", median_value) #printing the median value

```

```python
# 7. Plot latitude versus longitude and explain your observations.
plt.style.use('fivethirtyeight') # change the style or appearance of a plot
plt.scatter(df['longitude'], df['latitude'], s = 3) # using x and y as longitude & latitude with size of 3. 
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Latitude versus Longitude',color = 'orange')
plt.show()

#Scatter diagram: It is continuos data within a range and scatter diagram helped with the relationship between the x and y -axis.
#As per the graph clusters appear denser with data points, it indicates a concentration of data in those particular areas
# and low concentration in some areas which could indicate differnt land type - near ocean, etc. 
```

```python
# 8. Create a data set for which the ocean_proximity is â€˜Near oceanâ€™.

#filtering dataframe to create near_ocean_df that contains only the rows where the 'ocean_proximity' column is equal 
# to 'NEAR OCEAN'. 
near_ocean_df = df[df['ocean_proximity'] == 'NEAR OCEAN']
near_ocean_df
```

```python
# 9. Find the mean and median of the median income for the data set created in question 8.

#taking the near_ocean data set and using mean and median function on median income. (Statistical analysis)
mean_value = near_ocean_df['median_income'].mean()
print("Mean value : " , mean_value)

median_value = near_ocean_df['median_income'].median()
print("Median value: ", median_value)
```

```python
# 10. Please create a new column named total_bedroom_size. If the total bedrooms is 10 or less, it should be quoted as 
# small. If the total bedrooms is 11 or more but less than 1000, it should be medium, otherwise it should be considered 
# large

df['total_bedroom_size'] = df['total_bedrooms'].apply(lambda x: 'Small' if x <= 10 else ('medium' if 11 <= x < 1000 else 'large'))
#here new column total_bedroom_size is created where it is checking total bedrooms to get the output using the
#lambda function along with conditional statment
#data along with new column at the end can be seen. 
df
```
